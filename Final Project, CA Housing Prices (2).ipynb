```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
from sklearn import metrics
from sklearn.preprocessing import scale
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn import metrics
import matplotlib.pyplot as plt
```

First we properly scan the CA housing data .csv from the github repo

```python
df = pd.DataFrame()
df = pd.read_csv("https://raw.githubusercontent.com/Chris323/FileDump/main/california_house.csv")
#df.shape
```

Next we print the data to make sure it properly saved into a dataframe

```python
df[0::100]
```

Next we need to properly scale the data

```python
scaled_df = scale(df)
```

```python
scaled_df[0::100]
```

```python
df.columns.values
```

```python
df2 = pd.DataFrame(data = scaled_df, columns = df.columns.values)
```

```python
df2
```

Next we drop any nan values the data may contain

```python
df2.dropna(inplace=True)
df2
```

Next we create the x and y matrices, here is where I adjusted the feature sets to accomodate my tests.

```python
feature_columns = ['MedInc','HouseAge','AveRooms','AveBedrms','Population','AveOccup','Latitude','Longitude']
#feature_columns = ['HouseAge','AveRooms','AveBedrms','AveOccup'] #Performed Poorly, worst
#feature_columns = ['MedInc', 'Population','Latitude','Longitude'] # Second/Best
x = df2[feature_columns]
y = df2['SalePrice']
```

Here we check the data size and output

```python
print(x.shape)
x
```

```python
print(y.shape)
y
```

Next we create testing and training sets

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = 6)
```

Next we are going to create the final list that will contain all of the predictions for each prediction model. 

```python
final_dataset = []
final_dataset.clear()
```

Next we use the training and testing set on our prediction models, starting with Linear Regression

Linear Regression

```python
myLR = LinearRegression()
myLR.fit(x_train, y_train)
myLR_y_predict = myLR.predict(x_test)
myLR_mse = metrics.mean_squared_error(y_test, myLR_y_predict)
myLR_rmse = np.sqrt(myLR_mse)
final_dataset.append(myLR_y_predict)
#final_dataset #Checking proper append
#myLR_y_predict
```

Random Forest Regression

```python
myRF = RandomForestRegressor(n_estimators = 19, bootstrap = True, random_state = 3)
myRF.fit(x_train, y_train)
myRF_y_predict = myRF.predict(x_test)
myRF_mse = metrics.mean_squared_error(y_test, myRF_y_predict)
myRF_rmse = np.sqrt(myRF_mse)
final_dataset.append(myRF_y_predict)
#final_dataset #Checking proper append
#myRF_y_predict
```

Ada Boost Regression

```python
myADA = AdaBoostRegressor()
myADA.fit(x_train, y_train)
myADA_y_predict = myADA.predict(x_test)
myADA_mse = metrics.mean_squared_error(y_test, myADA_y_predict)
myADA_rmse = np.sqrt(myADA_mse)
final_dataset.append(myADA_y_predict)
#final_dataset
#myADA_y_predict
```

KNeighbors Regressor

```python
myKNR = KNeighborsRegressor()
myKNR.fit(x_train, y_train)
myKNR_y_predict = myKNR.predict(x_test)
myKNR_mse = metrics.mean_squared_error(y_test, myKNR_y_predict)
myKNR_rmse = np.sqrt(myKNR_mse)
final_dataset.append(myKNR_y_predict)
#final_dataset
#myADA_y_predict
```

Next we add in all our y predictions into one final dataframe, compute the RSME for the ave column, and compare the results.

```python
final_df = pd.DataFrame(final_dataset)
final_df = final_df.transpose()
final_df['Ave_SalePrice'] = final_df.mean(numeric_only = True, axis = 1)
final_y_predict = final_df['Ave_SalePrice'].values
final_df
```

Lastly, we compare the RMSE for all the columns.

```python
last_col_mse = metrics.mean_squared_error(y_test, final_y_predict)
last_col_rmse = np.sqrt(last_col_mse)
print('Column 0, Linear Regression RMSE:', myLR_rmse)
print('Column 1, Random Forest RMSE:', myRF_rmse)
print('Column 2, Ada Boost RMSE:', myADA_rmse)
print('Column 3, KNieghbors Regressor RMSE:', myKNR_rmse)
print('Column Ave_SalePrice, RMSE:', last_col_rmse)
```
