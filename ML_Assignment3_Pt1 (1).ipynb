```python
from  keras.wrappers.scikit_learn  import  KerasClassifier
from  sklearn.model_selection  import  GridSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
import pandas as pd
from keras.utils import np_utils
from keras.datasets import mnist
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

Here we import the data and print out a sliced version of the DataFrame

```python
df = pd.DataFrame()
df = pd.read_csv("https://github.com/mpourhoma/CS4662/raw/master/CTG_clean.csv")
df[0::100]
```

Next we scale the data and place it back into a DataFrame

```python
#scaler = preprocessing.StandardScaler().fit(df)
cols = df.columns.values
scaled_df = preprocessing.scale(df)
scaled_df
```

```python
df2 = pd.DataFrame(data = scaled_df, columns = df.columns.values)
df2
```

```python
cols
```

Next we create x and y matrices using the scaled data and the original labels

```python
feature_cols = ['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV',
                'MLTV', 'Width', 'Min', 'Max', 'Nmax', 'Nzeros', 'Mode', 'Mean', 
                'Median', 'Variance', 'Tendency']
x = df2[feature_cols]
y = df['NSP']

# print the first 5 rows
#x
y
```

We then split the data into testing and training sets

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)

# print the size of the traning set:
print(x_train.shape)
print(y_train.shape)

# print the size of the testing set:
print(x_test.shape)
print(y_test.shape)
```

Next we instantiate the ANN model

```python
my_ANN = MLPClassifier(hidden_layer_sizes=(30,), activation= 'logistic', 
                       solver='adam', alpha=1e-5, random_state=1, 
                       learning_rate_init = 0.02, verbose=True, tol=0.0001)
```

Next we use 10 fold cross validation to evaluate our model

```python
accuracy_list = cross_val_score(my_ANN, x_train, y_train, cv=10, scoring='accuracy')

print('\n\n','accuracy: ',accuracy_list)
```

```python
# use average of accuracy values as final result
accuracy_cv = accuracy_list.mean()

print(accuracy_cv)
```

Next we use our ANN on our testing set and report the accuracy.

```python
my_ANN.fit(x_train, y_train)
```

```python
y_predict_ann = my_ANN.predict(x_test)
print(y_predict_ann)
```

```python
score_ann = accuracy_score(y_test, y_predict_ann)
print('\n','accuracy: ',score_ann)
```

We then fix the random state for reproducibility 

```python
seed = 1  
np.random.seed(seed)
```

Next we use gridsearch for the best number of Neurons

```python
# define a range for the "number of neurons" in the hidden layer for a network with 1 hidden layer:
rangeList = list(range(5, 251))
```

```python
#rangeList[0::5]
```

```python
neuron_number = [(i,) for i in rangeList[0::5]]
```

```python
#neuron_number
```

```python
# create a dictionary for grid parameter:
param_grid = dict(hidden_layer_sizes = neuron_number)
print(param_grid,'\n')
```

```python
# instantiate the model:
my_ANN = MLPClassifier(hidden_layer_sizes=(30,), activation= 'logistic', 
                       solver='adam', alpha=1e-5, random_state=1, 
                       learning_rate_init = 0.02, verbose=True, tol=0.0001)
```

```python
# creat the grid, and define the metric for evaluating the model: 
grid = GridSearchCV(my_ANN, param_grid, cv=10, scoring='accuracy')
```

```python
# fit the grid (start the grid search):
grid.fit(x, y)
```

```python
# view the best results corresponding to the best structure of ANN:
print(grid.best_score_)
print(grid.best_params_)
```

The best accuracy score is 85% and the best number of neurons is 45 for the hidden layer
