```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
from sklearn import metrics
from sklearn.preprocessing import scale
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn import metrics
import matplotlib.pyplot as plt
```

Here, below, we import the data and place it in a DataFrame

Next we create a new DataFrame with only the numerical columns present

```python
numCol = df.select_dtypes(include=['number'])
numCol[0::100]
```

We make sure to note the shape of this new DataFrame, it will be needed later.

```python
numCol.shape
```

Next, we create an additional DataFrame with only the catagorical data.

```python
catCol = df.select_dtypes(include=[object])
catCol[0::100]
```

We also make note of this shape as well.

```python
catCol.shape
```

We made sure all the catagorical data were set as strings with the command below

```python
catCol = catCol.applymap(str)
```

We then use LabelEncoder to convert all our catagorical data into numerical values, and we place it in a new DF

```python
le = preprocessing.LabelEncoder()
df2 = catCol.apply(le.fit_transform)
```

```python
df2[0::100]
```

We then took our numerical df (numCol) we created, and the new (previously catagorical) df and merge the columns, creating a new df, in order to have all our data back as one.

```python
frames = [df2, numCol]

dfResult = pd.concat(frames, axis=1)
```

We then make sure our result df's shape matches our original df's shape.

```python
dfResult.shape
```

```python
dfResult[0::100]
```

```python
scaled_df = scale(dfResult)
```

```python
#scaled_df.columns.values
```

```python
dfFinal = pd.DataFrame(data = scaled_df, columns = dfResult.columns.values)
dfFinal = dfFinal.reset_index()
```

```python
feature_columns = dfFinal[dfFinal.columns.values]
x = feature_columns.drop(['SalePrice'], axis=1)
y = dfFinal['SalePrice']
```

```python
#x.shape
#y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state = 6)
```
