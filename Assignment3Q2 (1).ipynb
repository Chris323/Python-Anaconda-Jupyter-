```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
from sklearn import metrics
from sklearn.linear_model import LinearRegression
```

Importing, Reading, and Printing Data

```python
df = pd.DataFrame()
df = pd.read_csv("https://github.com/mpourhoma/CS4661/raw/master/Credit.csv")
df[0::10]
```

Here we label our data in an X and Y array

```python
feature_columns = ['Income','Limit','Rating','Cards','Age','Education','Married']
x = df[feature_columns]
y = df['Balance']
```

Checking X and Y

```python
print(x.shape)
print(y.shape)
```

```python
x[0::10]
```

```python
y[::10]
```

Here we normalize our X and print to check the results

```python
norm_x = preprocessing.scale(x) 
norm_x
```

Next we split our data into training and testing sets.

```python
x_train,x_test,y_train,y_test = train_test_split(x, y ,test_size = 0.24, random_state = 9)
```

Next we instantiate an object of the Linear Regression Class and fit the model to the training set.

```python
my_linReg = LinearRegression()
my_linReg.fit(x_train, y_train)
```

Here we print out the Theta

```python
print(my_linReg.intercept_)
print(feature_columns)
print(my_linReg.coef_)
```

"Cards" had the largest coefficient at 18.19349659, while " Married" had the lowest coefficient at -20.17447952

Next we Calculate the RMSE for the

```python
y_predict = my_linReg.predict(x_test)
mse = metrics.mean_squared_error(y_test, y_predict)
rmse = np.sqrt(mse)
#print(y_predict)
print("RMSE:", rmse)
```

Next we use 10-fold Cross Validation to evaluate our Linear Regression and print out the mean for our rmse list of values.

```python
mseList = cross_val_score(my_linReg, x, y, cv=10, scoring='neg_mean_squared_error')
positiveMseList= -mseList #get positive values
rmseList = np.sqrt(positiveMseList)
print("RMSE List Results:", rmseList)
print("RMSE List Results Mean:", rmseList.mean())
```
